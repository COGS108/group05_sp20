```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython.display import Image
import seaborn as sns
from scipy.stats import kstest
```

```python
# Configure libraries
# The seaborn library makes plots look nicer
sns.set()
sns.set_context('talk')

# Don't display too many rows/cols of DataFrames
pd.options.display.max_rows = 7
pd.options.display.max_columns = 8

# Round decimals when displaying DataFrames
pd.set_option('precision', 2)
```

```python
#Foodbank Data
df_foodbank = pd.read_csv('Foodbank.csv')
```

```python
#Has been sorted from most TOTAL LBS to least TOTAL LBS by CITY
df_foodbank
```

```python
df_foodbank.describe
```

```python
#Removing rows that have no value
df = df_foodbank[['PROGRAM','AGENCY #','CITY','ZIP CODE','FRESH PRODUCE BY LBS','TOTAL LBS']]
```

```python
df.describe
```

```python
df
```

```python
#Remove all variables that are not needed, such as 0s.
df = df[df['FRESH PRODUCE BY LBS'] != '0']
```

```python
df
```

```python
#REMOVE ALL ',' to make data more uniform
df = df.replace(',','', regex=True)
df
```

```python
df['TOTAL LBS'].describe()
```

```python
#REMOVE COLUMNS THAT HAVE NO SIGNIFICANCE.
df = df[['PROGRAM','ZIP CODE','TOTAL LBS']]
df
```

```python
#SINGLE OUT EACH DATA COLUMN
ZIP_CODE = df['ZIP CODE']
TOTAL = df['TOTAL LBS']
```

```python
data_dict = {'ZIP CODE':pd.Series(ZIP_CODE),'TOTAL LBS':pd.Series(TOTAL)}
```

```python
#Create individual dataframe from created dictionary
data = pd.DataFrame(data_dict)
```

```python
data
```

```python
#Convert information to floats

df = data.astype(float)
```

```python
#Plotting points with ZIP CODES and TOTAL LBS per ZIP CODE with scatterplot
fig = plt.figure(figsize=(100,200))
plt.xlabel("Zip Code")
plt.ylabel("Total LBS")
plt.title("TOTAL LBS by ZIP CODE Distribution")
plt.scatter(data['ZIP CODE'], data['TOTAL LBS'])
```
