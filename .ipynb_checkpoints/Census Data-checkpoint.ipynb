```python
import pandas as pd
df = pd.read_excel('~\Documents\GitHub\group05_sp20\Jan-Apr 2019 and 2020 by worksheet.xlsx')
df.head()
```

```python
len(df)
```

```python
zipcodes = df['ZIP CODE'].sort_values().unique()
zipcodes = zipcodes[:-2]
```

```python
df2 = pd.read_csv('~/Downloads/zip_code_database.csv')
df2.head()
```

```python
zipcodes_list = []
pop_estimates_2015 = []
for i in range(0, len(zipcodes)):
    zipcodes_list.append(zipcodes[i])
    pop_estimates_2015.append(df2[df2['zip']==zipcodes[i]]['irs_estimated_population_2015'].iloc[0])
pop_estimates_2015
```

```python
zipcodes_list = [int(i) for i in zipcodes_list]
zipcodes_list
```

```python
population = pd.DataFrame()
population['Zip Code'] = zipcodes_list
population['Population Estimate (2015)'] = pop_estimates_2015
population
```
