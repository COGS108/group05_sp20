```python
filepath = "Census_Data/Census_ZIP.shp"
tempmap = gpd.read_file(filepath)

pop = pd.read_excel("Population_Estimate.xlsx").groupby('ZIP')['POPULATION'].sum().to_frame()
sdmap= pd.merge(tempmap, pop, on='ZIP', how='left')
```

```python
#Read in all foodbank data
#2019
Jan2019 = pd.read_excel("San_Diego_Foodbank_Data/Jan 2019.xlsx")
Feb2019 = pd.read_excel("San_Diego_Foodbank_Data/Feb 2019.xlsx")
Mar2019 = pd.read_excel("San_Diego_Foodbank_Data/Mar 2019.xlsx")
Apr2019 = pd.read_excel("San_Diego_Foodbank_Data/Apr 2019.xlsx")
#2020
Jan2020 = pd.read_excel("San_Diego_Foodbank_Data/Jan 2020.xlsx")
Feb2020 = pd.read_excel("San_Diego_Foodbank_Data/Feb 2020.xlsx")
Mar2020 = pd.read_excel("San_Diego_Foodbank_Data/Mar 2020.xlsx")
Apr2020 = pd.read_excel("San_Diego_Foodbank_Data/Apr 2020.xlsx")

#Note: Number of foodbanks sampled varies from year to year and month to month.
#----------------------------------------------------------------------------------
    # Jan 2019: 452
    # Feb 2019: 387
    # Mar 2019: 431
    # Apr 2019: 418
    #--------------
    # Jan 2020: 486
    # Feb 2020: 331
    # Mar 2020: 481
    # Apr 2020: 454
#----------------------------------------------------------------------------------
# We elect to consider only foodbanks that consistently appear throughout all dfs.

def get_df_name(df):
    name =[x for x in globals() if globals()[x] is df][0]
    return name

food_banks = {}
data = [Jan2019, Jan2020, Feb2019, Feb2020, Mar2019, Mar2020, Apr2019, Apr2020]
for i in range(len(data)):
    for j, row in data[i].iterrows():
        if row['AGENCY #'] not in food_banks:
            if i > 0:
                continue
            food_banks[row['AGENCY #']] = { 'zip' : row['ZIP CODE'], 'months' : {} }
        food_banks[row['AGENCY #']]['months'][get_df_name(data[i])] = round(row['TOTAL LBS'])
        
#food_banks(map(lambda x: len(x['months']), food_banks))
food_banks = { k: v for k, v in food_banks.items() if len(v['months']) == len(data) }
```

* As previously mentioned, the number of observations throughout the data set provided by the San Diego Food Bank varies from sheet to sheet. We elected to focus our efforts on a subset of the foodbank data which contains only agencies who have obvservation throughout all 8 sheets, hoping this would enable us to draw more concrete conclusions about the data. 
* The information extracted was then organized into a dictionary of dictionaries, later used to generate other dataframes.

```python
food = {}
for month in data:
    month_name = get_df_name(month)
    for k, v in food_banks.items():
        if month_name not in food:
            food[month_name] = { v['zip']: 0 }
        elif v['zip'] not in food[month_name]:
            food[month_name][v['zip']] = 0
        food[month_name][v['zip']] += v['months'][month_name]      
food

df_jan2019 = pd.DataFrame(columns=['ZIP', 'food_distributed'])
for k,v in food['Jan2019'].items():
    newrow = {'ZIP':int(k), 'food_distributed':v}
    df_jan2019 = df_jan2019.append(newrow, ignore_index=True)
df_jan2019

df_jan2020 = pd.DataFrame(columns=['ZIP', 'food_distributed'])
for k,v in food['Jan2020'].items():
    newrow = {'ZIP':int(k), 'food_distributed':v}
    df_jan2020 = df_jan2020.append(newrow, ignore_index=True)
df_jan2020

df_feb2019 = pd.DataFrame(columns=['ZIP', 'food_distributed'])
for k,v in food['Feb2019'].items():
    newrow = {'ZIP':int(k), 'food_distributed':v}
    df_feb2019 = df_feb2019.append(newrow, ignore_index=True)
df_feb2019

df_feb2020 = pd.DataFrame(columns=['ZIP', 'food_distributed'])
for k,v in food['Feb2020'].items():
    newrow = {'ZIP':int(k), 'food_distributed':v}
    df_feb2020 = df_feb2020.append(newrow, ignore_index=True)
df_feb2020

df_mar2019 = pd.DataFrame(columns=['ZIP', 'food_distributed'])
for k,v in food['Mar2019'].items():
    newrow = {'ZIP':int(k), 'food_distributed':v}
    df_mar2019 = df_mar2019.append(newrow, ignore_index=True)
df_mar2019

df_mar2020 = pd.DataFrame(columns=['ZIP', 'food_distributed'])
for k,v in food['Mar2020'].items():
    newrow = {'ZIP':int(k), 'food_distributed':v}
    df_mar2020 = df_mar2020.append(newrow, ignore_index=True)
df_mar2020

df_apr2019 = pd.DataFrame(columns=['ZIP', 'food_distributed'])
for k,v in food['Apr2019'].items():
    newrow = {'ZIP':int(k), 'food_distributed':v}
    df_apr2019 = df_apr2019.append(newrow, ignore_index=True)
df_apr2019

df_apr2020 = pd.DataFrame(columns=['ZIP', 'food_distributed'])
for k,v in food['Apr2020'].items():
    newrow = {'ZIP':int(k), 'food_distributed':v}
    df_apr2020 = df_apr2020.append(newrow, ignore_index=True)
```

* The data was organized into dataframes so it could then by merged with the population information.

```python
def adjust(df):
    adjusted_df = pd.merge(df, pop, on='ZIP', how='left')
    adjusted_df = adjusted_df[adjusted_df['POPULATION'] != 0]
    adjusted_df = adjusted_df.assign(ratio = adjusted_df['food_distributed'] / adjusted_df['POPULATION'])
    adjusted_df['ZIP'] = pd.to_numeric(adjusted_df['ZIP'])
    adjusted_df['food_distributed'] = pd.to_numeric(adjusted_df['food_distributed'])
    adjusted_df['ratio'] = pd.to_numeric(adjusted_df['ratio'])
    return adjusted_df
```

* adjust(df) creates an adjusted dataframe, containing a column called ratio - a calculation of food_distributed/population.

```python
adjusted_df_jan2019 = adjust(df_jan2019)
adjusted_df_feb2019 = adjust(df_feb2019)
adjusted_df_mar2019 = adjust(df_mar2019)
adjusted_df_apr2019 = adjust(df_apr2019)
adjusted_df_jan2020 = adjust(df_jan2020)
adjusted_df_feb2020 = adjust(df_feb2020)
adjusted_df_mar2020 = adjust(df_mar2020)
adjusted_df_apr2020 = adjust(df_apr2020)
```

* adjust(df) is then called on each of the monthly data frames.

```python
ratios1 = adjusted_df_jan2019.drop(['ZIP', 'food_distributed', 'POPULATION'], axis=1)
ratios2 = adjusted_df_feb2019.drop(['ZIP', 'food_distributed', 'POPULATION'], axis=1)
ratios3 = adjusted_df_mar2019.drop(['ZIP', 'food_distributed', 'POPULATION'], axis=1)
ratios4 = adjusted_df_apr2019.drop(['ZIP', 'food_distributed', 'POPULATION'], axis=1)
ratios5 = adjusted_df_jan2020.drop(['ZIP', 'food_distributed', 'POPULATION'], axis=1)
ratios6 = adjusted_df_feb2020.drop(['ZIP', 'food_distributed', 'POPULATION'], axis=1)
ratios7 = adjusted_df_mar2020.drop(['ZIP', 'food_distributed', 'POPULATION'], axis=1)
ratios8 = adjusted_df_apr2020.drop(['ZIP', 'food_distributed', 'POPULATION'], axis=1)
```

```python
df_ethnicity = pd.read_excel('Ethnicity Zip.xlsx')
```

```python
ethaNA = df_ethnicity[df_ethnicity['ETHNICITY']=='American Indian'] #Native American
ethA = df_ethnicity[df_ethnicity['ETHNICITY'] == 'Asian']
ethB = df_ethnicity[df_ethnicity['ETHNICITY'] == 'Black']
ethL = df_ethnicity[df_ethnicity['ETHNICITY'] == 'Hispanic'] #Latinx
ethO = df_ethnicity[df_ethnicity['ETHNICITY'] == 'Other']
ethPI = df_ethnicity[df_ethnicity['ETHNICITY'] == 'Pacific Islander']
ethTM = df_ethnicity[df_ethnicity['ETHNICITY'] == 'Two or More']
ethW = df_ethnicity[df_ethnicity['ETHNICITY'] == 'White']
```

* df_ethnicity is partioned into seperate dataframes, each representing a different ethnic grouping.

```python
pop['POPULATION'] = pop['POPULATION'].replace(0,-1)
```

* 0 values are replaced with -1, so that the proportions can be calculated in the prop function defined below:

```python
def prop(df):
    props = []
    for i in range(len(df)):
        zipc = df.iloc[i]['ZIP']
        if zipc in pop.index:
            props.append(df.iloc[i]['POPULATION']/pop.loc[zipc][0])
        else:
            continue
    df['PROPORTION'] = props
    df[df['PROPORTION'] < 0] = 0
```

* Note that negative values are then adjusted to prevent skewing the analyses.

```python
prop(ethaNA)
prop(ethA)
prop(ethB)
prop(ethL)
prop(ethO)
prop(ethPI)
prop(ethTM)
prop(ethW)
```

* prop is applied to each ethnic group.

* Now we procceed to group together the common months, shuffle the ratios, and then re-calculate the difference of means of the months.

```python
def month_bootstrap(df1, df2):
    df1['year'] = 2019
    df2['year'] = 2020
    month = pd.concat([df1, df2])
    changes = []
    for i in range(1000):
        month['ratio'] = np.random.permutation(month['ratio'].values)
        changes.append(month[month['year']==2020].ratio.mean() - month[month['year']==2019].ratio.mean())
    return pd.DataFrame({'meandiff': changes})
```

* The function concatenates two data sets of the same length, concatenates them, randomizes the ratio column and calculates the difference of means between 2020 and 2019 a thousand times to generate a distribution of the sample difference of means.

```python
diff = ratios5.ratio.mean() - ratios1.ratio.mean()
jan_bootstrap = month_bootstrap(ratios1, ratios5)
jan_bootstrap.meandiff.hist()
plt.axvline(diff, color='C1')
```

* The orange vertical line represents the orignal difference of means between january 2019 and january 2020, and the distribution represents the 1000 randomized differences of means.

```python
jan_bootstrap.meandiff.quantile(0.025), jan_bootstrap.meandiff.quantile(0.975)
```

* The confidence interval shows that our original difference of means falls within the expected range. This process is then repeated for Feb-Apr.

```python
#February
diff = ratios6.ratio.mean() - ratios2.ratio.mean()
feb_bootstrap = month_bootstrap(ratios2, ratios6)
feb_bootstrap.meandiff.hist()
plt.axvline(diff, color='C1')
```

```python
feb_bootstrap.meandiff.quantile(0.025), feb_bootstrap.meandiff.quantile(0.975)
```

```python
#March
diff = ratios7.ratio.mean() - ratios3.ratio.mean()
mar_bootstrap = month_bootstrap(ratios3, ratios7)
mar_bootstrap.meandiff.hist()
plt.axvline(diff, color='C1')
```

```python
apr_bootstrap.meandiff.quantile(0.025), apr_bootstrap.meandiff.quantile(0.975)
```

* By grouping data of the same month (to track a year-to-year change), randomizing the ratios, and bootstrapping the difference in means across years, we see that, for each month, the yearly change falls within values we would expect them to for this type of data, which is also confirmed by their respective 95% confidence intervals. Therefore, we feel confident that in a situation like COVID-19, the difference in means of ratios we observe over a year would be similar to the results we have.

### Ethics & Privacy

The data sets we used, namely the San Diego Food Bank distribution data, and SANGIS and SANDAG data, consisted of aggregate data and contained no personal identifiers, so privacy was not a concern. The data we were able to obtain contained estimates of the total population and the ethnic makeup of San Diego County zip codes, alongside information regarding the amount of food distributed by a subset of the San Diego Foodbank's agencies. The lack of consistent food distribution information for all of San Diego county's zip codes adversely affected our ability to generalize or conclusions, and say with confidence that our work is empirically valid. Furthermore, our failure to track other correlates, such as gender or median income due to a lack of time and access to sufficiently granular data. By not tracking these correlates, might very well have resulted in skewed results and biased interpretations.

It's also worth mentioning that the San Diego Food Bank is not the only hunger-relief organization operating in San Diego County. Other organizations were not taken into account due to a lack of available data. Even if we had access to other organizations' data, underserved communities and individuals might have their food needs met by organizations and individuals outside the scope of available data, (churches, religious institutions), might elect not to participate in a census, and/or be documented. 

The ethical implications of drawing conclusions based on our research without considering the above are serious. If our work were taken into account by decision-makers and stakeholders, it might skew the distribution of food, adversely affecting communities in need that were not considered, and possibly exacerbating the situation. In light of this, we urge our readers to consider the limitations of our work when drawing conclusions about how the COVID-19 pandemic has affected food distribution.

### Conclusion & Discussion

Our work centered around determining how the outbreak of COVID-19 has affected the amount of food measured in LBs distributed by the San Diego Food Bank, particularly in communities of color and among those of limited financial means. When we formulated our hypothesis, we predicted that the outbreak of COVID-19 has negatively impacted individuals' access to food, disproportionately affecting the financially insecure and communities of color. The COVID-19 outbreak is ongoing, and our ability to collect and analyze sufficient amounts of information to confirm or reject our null hypothesis was limited. Although we were able to show that there has been a significant increase in the amount of food distributed across the zip codes we tracked, and provide contextual evidence that indicated that certain communities were disproportionately affected, we were ultimately unable to empirically determine whether the outbreak of the pandemic has disproportionately affected communities of color and the financially insecure.

As mentioned in the Ethics & Privacy section, our work is extremely limited. Any future research into the subject would benefit from access to more granular data about unemployment, (preferably by zip code), and more consistent information regarding food distribution throughout all the zip codes in San Diego county by all of the hunger-relief organizations operating in the area. Furthermore, future work would benefit from taking into consideration possible correlates through the inclusion of additional demographic information regarding the gender and median income. Doing so might enable researchers to obtain a deeper understanding of the phenomenon, and draw more concrete conclusions.
