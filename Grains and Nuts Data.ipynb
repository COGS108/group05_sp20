```python
import pandas as pd
df = pd.read_csv('~/Downloads/Field Crop Data (2019-2020).csv')
df
```

```python
df = df[['Year', 'Period', 'Commodity', 'Data Item', 'Value']]
df
```

```python
df = df[df['Value'] != ' (D)']
df = df[df['Value'] != ' (S)']
df = df[df['Value'] != ' (NA)']
df
```

```python
df = df.reset_index(drop=True)
df = df[['Year', 'Period', 'Commodity', 'Data Item', 'Value']]
df
```

```python
df2019 = df[df['Year'] == 2019]
df2019
```

```python
df2020 = df[df['Year'] == 2020]
df = pd.concat([df2019, df2020])
df
```
